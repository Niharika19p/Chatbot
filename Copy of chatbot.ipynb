```python
import re
def simple_chatbot():
    print("ChatBot: Hello! How can I help you today? (type 'bye' to exit)")
    while True:
        user_input = input("You: ").lower()
        if user_input == 'bye':
            print("ChatBot: Goodbye! Have a great day!")
            break
        elif re.search(r'\bhi\b|\bhello\b|\bhey\b', user_input):
            print("ChatBot: Hi there! How can I assist you?")
        elif re.search(r'\bhow are you\b', user_input):
            print("ChatBot: I'm good. what about you?")
        elif re.search(r'\bgood\b|\bgreat\b|\bnice\b', user_input):
            print("ChatBot: that's great!, How can i help you?")
        elif re.search(r'\bwhat is your name\b', user_input):
            print("ChatBot: I'm a simple chatbot created to help you.")
        elif re.search(r'\bhelp\b|\bsupport\b', user_input):
            print("ChatBot: Sure, I can help. Please tell me your problem.")
        elif re.search(r'\bthank you\b|\bthanks\b', user_input):
            print("ChatBot: You're welcome!")
        elif re.search(r'\bsuggest me some movies\b', user_input):
            print("ChatBot: Here some movies for you: Train to busan,The mermaid,Gangnam zombie,Memoir of a murder")
        elif re.search(r'\bgoodmorning\b', user_input):
            print("ChatBot: Good morning!")
        elif re.search(r'\bgoodnight\b', user_input):
            print("ChatBot: Good night!")
        else:
            print("ChatBot: I'm sorry, I didn't understand that.")


simple_chatbot()
```
